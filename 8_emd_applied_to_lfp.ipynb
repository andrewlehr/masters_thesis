# Applying EMD to hippocampal LFP
## Produces Figures 30-35
EMD is applied to LFP data. We have a look at the IMFs and see there is apparent mode mixing. We assess phase-amplitude coupling between the IMF that best corresponds to the 8 Hz theta rhythm and IMFs in the gamma band.

### Import Libraries

```python
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import numpy as np
import scipy
from scipy.io import loadmat
import PyEMD
```

### Load data

```python
x = loadmat('JBCA1ppDG_1-240615-SOR-open_field_egf2_ex.mat')
x = x['egf']
x = x.reshape(x.shape[0])
x = x.astype('float64')
Fs = 4800
```

### Downsample

```python
factor = 1 #6
if factor > 1:
    Fs = Fs/factor
    x = scipy.signal.decimate(x, q=factor, zero_phase=True)
```

### Convert signal to micro or millivolt

```python
adc_conv = 1500/1000                      # ADC_fullscale_mv 1500
gain = 2000                               # gain_ch_28 2000
max_int_val = 2**(2*8-1)                  # 2 byte integers
R = 10000                                 # 10kOhm attenuator (in 0_Protocol - CA1ppDG Extinction Study.rtf)
attenuation = (R + 2000) / 2000           # annenuation factor

factor = attenuation/max_int_val*adc_conv/gain;  # from mTintCore read_eeg_file.m
x = x*factor                                     # converts to volt
x = 1000*1000*x                                  # converts to microvolt
```

### Extract IMFs

```python
# Define data subsample, nSeconds
nSeconds = 20. # 120
X = x[0:int(Fs*nSeconds)]
t = np.arange(0, nSeconds, 1/Fs)
```

```python
plt.figure(figsize=(30,10))
plt.plot(X, linewidth=3)
plt.xlim(0, Fs)
plt.show()
```

```python
# Extract imfs and residue
emd = PyEMD.EMD()
emd.emd(X)
imfs, res = emd.get_imfs_and_residue()
nIMF = imfs.shape[0]
```

### Compute analytic signal and instantaneous frequency

```python
analytic_signal = scipy.signal.hilbert(imfs)
amplitude_envelope = np.abs(analytic_signal)
instantaneous_phase = np.unwrap(np.angle(analytic_signal))
instantaneous_frequency = (np.diff(instantaneous_phase) /
                           (2.0*np.pi) * Fs)
```

### Compute mean frequencies of IMFs

```python
np.set_printoptions(suppress=True)
f_m = np.round(np.mean(instantaneous_frequency,1),2)
```

### Compute standard deviation of instantaneous frequency

```python
nT = np.size(instantaneous_frequency, axis=1)
n_perc = 20
perc = 0.0001
inst_freq_std = np.zeros(shape=(nIMF,))
s1 = int(perc * nT)
s2 = -int(perc * nT)
inst_freq_std = np.std(instantaneous_frequency[:, s1:s2], axis=1)
```

### Print instantaneous frequency and std

```python
for i in range(len(inst_freq_std)):
    print('IMF ' + str(i) + ' | Frequency: ' + str(f_m[i]) + ', std: ' + str(np.round(inst_freq_std[i], 2)))
```

### Power spectrum of imfs

```python
nfft = len(X)
s1 = 0           # defines start of frequency window
s2 = 35          # defines end of frequency window

#colors = ['chocolate', 'forestgreen', 'saddlebrown', 'slategray']
colors = ['dimgrey', 'black', 'grey', 'lightslategrey']
l_style = [(0, (3, 1, 1, 1, 1, 1)), 'solid', 'dashed', 'dotted']

fig, ax = plt.subplots(figsize=(15,9))
for k in range(6, nIMF-4):
    # compute fft and get frequencies
    x_hat = np.fft.fft(imfs[k,:], nfft)
    timestep = 1/Fs
    freq = np.fft.fftfreq(nfft, d=timestep)

    # select only freq interval [s1, s2] to display power spectrum
    f_sel = freq[(s1<freq)&(freq<s2)]

    # compute periodogram (absolute value of fourier transform divided by number of points)
    periodogram = (1/nfft) * np.abs(x_hat[(s1<freq)&(freq<s2)])
    log_periodogram = np.log10(periodogram)

    # smooth the periodogram
    n_freqs = len(f_sel)                             # number of frequency bins
    n_pts = 5                                        # window, two-sided  
    periodogram_smooth = np.zeros(n_freqs - 2*n_pts) # length is 2*n_pts shorter than periodogram
    for t_i in range(n_pts, n_freqs-n_pts):
        periodogram_smooth[t_i-n_pts] = np.mean(periodogram[(t_i-n_pts) : (t_i+n_pts)])

    # strongest frequency
    max_f = f_sel[n_pts + np.argmax(periodogram_smooth)]

    plt.plot(f_sel[n_pts:(n_freqs-n_pts)], periodogram_smooth, color=colors[k-6], linestyle=l_style[k-6], label=np.round(f_m[k],2))#, color='black')
    
plt.xlim(s1, s2)
plt.xticks(fontsize=20)
plt.xlabel('Frequency (Hz)', fontsize=24)
#plt.ylim(0, 65)
plt.yticks(fontsize=20)
plt.ylabel('Power ' + r'($\frac{\mu V^2}{N}$)', fontsize=24)


handles, labels = ax.get_legend_handles_labels()
leg = ax.legend(reversed(handles), reversed(labels), loc='upper right', fontsize=16,  handlelength=2.65)  # reverse to keep order consistent
leg_lines = leg.get_lines()
plt.setp(leg_lines, linewidth=4)

plt.savefig('plots/ch5-imfs_power_spec.pdf', bbox_inches = "tight")

plt.show()
```

### Phase-Amplitude Coupling

#### Compute for theta, 8Hz

```python
# phase-amplitude coupling
ex = 7
ipm = instantaneous_phase[ex,:] % (2*np.pi)
bins = np.linspace(0, 2*np.pi, 36)
count = np.zeros((len(bins)-1,nIMF))
for i in range(nIMF):
    for j in range(1,len(bins)):
        inBin = (ipm > bins[j-1]) & (ipm < bins[j])
        count[j-1,i] = np.mean(amplitude_envelope[i, inBin])
    count[:,i] = count[:,i]/np.max(count[:,i])
    
# ex amplitude   
amp = np.zeros((len(bins)-1,1))
for j in range(1,len(bins)):
    inBin = (ipm > bins[j-1]) & (ipm < bins[j])
    amp[j-1] = np.mean(imfs[ex,inBin])
```

```python
fig, ax = plt.subplots(figsize=(12,8))

print(str(f_m[ex]))
bin_centers = np.degrees(bins[1:] + (bins[1:]-bins[:-1])/2)

ax = plt.subplot(211)
for j in range(3, nIMF-3):
    plt.plot(bin_centers, count[:,j], linewidth=1, label=str(f_m[j]))

plt.xlim(bin_centers[0], bin_centers[-1])
plt.xticks([])
plt.yticks(fontsize=14)
plt.ylabel('Normalized Envelope', fontsize=16)

handles, labels = ax.get_legend_handles_labels()
leg = ax.legend(reversed(handles), reversed(labels), loc='lower right', fontsize=12,  handlelength=2)  # reverse to keep order consistent
leg_lines = leg.get_lines()
plt.setp(leg_lines, linewidth=2 )

plt.subplot(212)

plt.plot(bin_centers, amp, linewidth=1, linestyle='-', color='black')
plt.xlim(bin_centers[0], bin_centers[-1])
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Theta Phase (degrees)', fontsize=16)
plt.ylabel('Average Voltage ($\mu V$)', fontsize=16)

fig.subplots_adjust(hspace=0.1)
fig.align_ylabels()

plt.savefig('plots/ch5_theta_gamma.pdf', bbox_inches = "tight")

plt.show()
```
